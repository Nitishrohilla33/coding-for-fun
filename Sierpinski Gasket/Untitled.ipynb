```python
import numpy as np
import matplotlib.pyplot as plt

# Vertices of triangle
vertices = np.array([
    [0.0, 0.0],
    [1.0, 0.0],
    [0.5, np.sqrt(3)/2]
])

# Initial point (no need for complicated geometry)
point = np.random.rand(2)

steps = 1000000

# Preallocate (faster than append)
points = np.zeros((steps, 2))

for i in range(steps):
    v = vertices[np.random.randint(0, 3)]
    point = (point + v) / 2
    points[i] = point

# Plot
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], color='black', marker = ',', linestyle = ' ',markersize = 0.8)
plt.axis('equal')
plt.axis('off')  # cleaner look
plt.title('Sierpinski Gasket')

#plt.savefig('sierpinski.png', dpi=600, bbox_inches='tight')
plt.show()
```
